# Exercise: Name and company

1. Ask the user their name and company, and assign to two variables.
2. Answer in one of four ways:
    - If both are the same as you, say "You must be me!"
    - If the name is the same, but the company is different, say, "Nice name, bad company"
    - If the company is the same, but the name is different, then say, "Hi my colleague"
    - If neither is the same, then say something rude about their name + company.

```python
name = input('Enter your name: ')
company = input('Enter your company: ')

my_name = 'Reuven'
my_company = 'Lerner'

if name == my_name and company == my_company:
    print('You must be me!')
elif name == my_name and company != my_company:
    print('Great name, terrible company!')
elif name != my_name and company == my_company:
    print('Hello, my colleague!')
else:
    print('Your name and company are both awful!')
    
```

# Exercise: Pig Latin 

- Get a word from the user, check if it starts with a, e, i, o, u (vowel)
- If so, add `way` to the word
- If not, move the first letter to the end, and add `ay`

Examples:
- `computer` -> `omputercay`
- `table` -> `abletay`
- `papaya` -> `apayapay`
- `elephant` -> `elephantway`
- `ice` -> `iceway`

```python
word = input('Enter a word: ')

if word[0] == 'a' or word[0] == 'e' or word[0] == 'i' or word[0] == 'o' or word[0] == 'u':
    print(word + 'way')
```

```python
word = input('Enter a word: ')

#    True/False   or True or True or True or True:
if word[0] == 'a' or 'e' or 'i' or 'o' or 'u':
    print(word + 'way')
```

```python
word = input('Enter a word: ')

if word[0] in 'aeiou':
    print(word + 'way')
else:
    print(word[1:] + word[0] + 'ay')
```

# Exercise: Find a character

1. Ask the user to enter a text string.
2. Ask the user to enter a numeric index.
3. Print the character and string and index.
4. BUT if the index isn't numeric, scold the user
5. If the index is too low (< 0), scold the user
6. If the index is too high, scold the user

Example:

    Enter text: hello
    Enter index: 1
    index 1 in hello is e

    Enter text: hello
    Enter index: 100
    index 100 is too big!

```python
text = input('Enter text: ').strip()
index = input('Enter index: ').strip()

if index.isdigit():
    index = int(index)

    if index < 0:
        print(f'Index {index} is too small!')
    elif index >= len(text):
        print(f'Index {index} is too big!')
    else:
        print(f'Index {index} in {text} is {text[index]}')
else:
    print(f'{index} is not numeric!')
```

# `for` loops

- `for` - starts the `for` loop
- Then we have the loop variable
- Then `in`
- Then we have, at the end of the line, the value we're iterating over
- Then `:` -- at the end of the line
- Then, indented, the loop body

```python
text = 'abcd'

for one_character in text[2:5]:
    print(one_character)
```
