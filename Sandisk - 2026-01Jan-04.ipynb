# Two modes for cell input

- Edit mode -- what I type goes into the cell. I can enter this mode with ENTER or by clicking in the cell
- Command mode -- this gives Jupyter commands. I can get into command mode by clicking to the left or pressing ESC.

## Commands

- `c` -- copy the current cell
- `x` -- cut the current cell
