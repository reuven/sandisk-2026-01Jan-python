# `if`

- `if` looks right, and asks: Do I have `True` or `False`?
- At the end of the line, we have `:`
- The block is indented, usually with 4 spaces
- The block can be as long/short as you want
- `if` or `else` block runs, but not zero and not both

```python
name = input('Enter your name: ')

if name == 'Reuven':
    print('Hi, boss!')
    print('Nice to see you!')
elif name == 'xyz':
    print('That is a weird name!')
else:
    print('Hello, ' + name + ', whoever you are.')
```

# Combining conditions 

- `and`
- `or`
- `not`

```python
x = 10
y = 20

#    True  and    True  -> True
if x == 10 and y == 20:
    print('Both are what you want!')
```

```python
x = 10
y = 20

#    True  and    False  -> False
if x == 10 and y == 99:
    print('Both are what you want!')
```

```python
x = 10
y = 20

#    True  or    True  -> True
if x == 10 or y == 20:
    print('At least one is what you want!')
```

```python
x = 10
y = 20

#    True  or    False  -> True
if x == 10 or y == 90:
    print('At least one is what you want!')
```

# Exercise: Name and company

1. Ask the user their name and company, and assign to two variables.
2. Answer in one of four ways:
    - If both are the same as you, say "You must be me!"
    - If the name is the same, but the company is different, say, "Nice name, bad company"
    - If the company is the same, but the name is different, then say, "Hi my colleague"
    - If neither is the same, then say something rude about their name + company.

```python
name = input('Enter your name: ')
company = input('Enter your company: ')

my_name = 'Reuven'
my_company = 'Lerner'

if name == my_name and company == my_company:
    print('You must be me!')
elif name == my_name and company != my_company:
    print('Great name, terrible company!')
elif name != my_name and company == my_company:
    print('Hello, my colleague!')
else:
    print('Your name and company are both awful!')
    
```

# Boolean context

Everything to the right of `if` (and other boolean contexts) is `True` except for:

- `False`
- `None`
- 0
- Every empty type -- `''`, `[]`, `()`, `{}`

```python
name = input('Enter your name: ')

if name == '':   # not Pythonic! 
    print('Hey! You did not enter a name!')
else:
    print('Hello, ' + name + '.')
```

```python
name = input('Enter your name: ')

if name:   # not an empty string...
    print('Hello, ' + name + '.')
else:      # must be the empty string!
    print('Hey! You did not enter a name!')
```

```python
name = input('Enter your name: ')

if not name:
    print('You did not enter a name! ')
```
